```python
#import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
#import datetime as dt
```

```python
nRowsRead = None
df1 = pd.read_csv('kaggle\input\movilidad-de-personas-sample.csv', delimiter=',', nrows = nRowsRead, index_col = False)
df1.dataframeName = 'movilidad-de-personas-sample.csv'
nRow, nCol = df1.shape
print(f'Hay {nRow} de filas y {nCol} columnas')
```

```python
print('Tipos de datos de las columnas:\n', df1.dtypes)
print ('Número de datos que contiene:\n', df1.size)
print('Nombres de columnas:', df1.columns)
print('Nombres de filas:', df1.index)
```

```python
#Filtrar las columnas del DataFrame para quedarse con las siguientes 13 columnas

#date, destination_activity, destination_autonomous_region_name, destination_province_name, distance_km_interval, home_province_name, origin_activity,
#origin_autonomous_region_name, origin_province_name, total_km_traveled, year, month, day 

# fecha, actividad de destino, nombre de la región autónoma de destino, nombre de la provincia de destino, intervalo de km de distancia, nombre de la provincia de origen,
#actividad de origen, nombre de la región autónoma de origen, nombre de la provincia de origen, total de km recorridos, año, mes, día

# Purgamos hasta quedarnos sólo con 7 columnas:
"""
fecha ---------------------> date
motivo_destino ------------> destination_activity
comunidad_destino ---------> destination_autonomous_region_name
intervalo_kms_distancia ---> distance_km_interval
motivo_origen -------------> origin_activity
comunidad_origen ----------> origin_autonomous_region_name
total_kms -----------------> total_km_traveled
"""
# Más adelante podemos/debemos incluir year, month, day para mejorar las búsquedas
columnas = ['date', 'destination_activity', 'destination_autonomous_region_name', 'distance_km_interval', 'origin_activity',
            'origin_autonomous_region_name', 'total_km_traveled' ]
#df2 = df1
#columnas.extend([col for col in df2])
#df2 = df2[columnas]
#df2.head(5)
df2 = df1[columnas]
df2.head(5)
```

```python
df3 = df2.rename (columns = {'date' : 'fecha',
                       'destination_activity' : 'motivo_destino',
                       'destination_autonomous_region_name' : 'comunidad_destino',
                       'distance_km_interval' : 'intervalo_kms_distancia',
                       'origin_activity' : 'motivo_origen', 
                       'origin_autonomous_region_name' : 'comunidad_origen',
                       'total_km_traveled' : 'total_kms'})
```

```python
#Buscamos nulls y los contabilizamos. En este caso no es necesario borrarlos porque no hay.
df3.isnull().sum()
```

```python
madrid_d = df3['comunidad_destino'] == 'Madrid, Comunidad de'
madrid_d.head()
df4 = df3[madrid_d]
df4.head(10)
```

```python
print('Dimensiones:', df4.shape)
print('Número de elementos:', df4.size)
```

```python
madrid_o = df3['comunidad_origen'] == 'Madrid, Comunidad de'
madrid_o.head()
df5 = df3[madrid_o]
df5.head(10)
```

```python
print('Dimensiones:', df5.shape)
print('Número de elementos:', df5.size)
```

```python
madrid_total_desplazamientos = pd.concat ([df4, df5], ignore_index = True).drop_duplicates()
print('Dimensiones:', madrid_total_desplazamientos.shape)
print('Número de elementos:', madrid_total_desplazamientos.size)
```
